# Wordopt:
#### Removing the noise/not important  from the news column {"text"} so that modle can understand the only vital stuff

```python
# importing the Regular Expression -- used for Data cleaning or manipulation of the data
import re
```

```python
#Defining the function
def wordopt(text):
  #convert into lowercase
  text = text.lower()

  #remove punctution , URL
  text = re.sub(r'https?: //\S+|www\.\S+' , '' , text)
  #remove HTML Tags
  text = re.sub(r'<.*?>' , '' , text)
  #remove punctuation
  text = re.sub(r'[^\w\s]' , '' , text)
  #remove digits
  text = re.sub(r'\d' , '' , text)
  #remove newline characters
  text = re.sub(r'\n' , '' , text )
  return text
```

```python
# applying it on the Text column in the News dataset to clean the data
news['text'] = news['text'].apply(wordopt)
```

# Building the Predictive system

```python
# Defining the output when a news prediction come
def output_label(n):
  if n == 0:
    return 'It is a Fake News'
  elif n ==1:
    return 'It is a Genuine News'
```

```python
#defining the overall functionality that are describe above
def manual_testing(news):
  testing_news = {"text" : [news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_X_test = new_def_test["text"]
  new_X_test_vec = vectorization.transform(new_X_test)
  pred_LR = LR_model.predict(new_X_test_vec)
  pred_DTC = DTC_model.predict(new_X_test_vec)
  pred_RFC = RFC_model.predict(new_X_test_vec)
  predict_GBC = GBC_model.predict(new_X_test_vec)
  return "\n\nLR Prediction: {} \n\nDTC Prediction: {} \n\nRFC Prediction: {} \n\nGBC Predicion : {}" . format(output_label(pred_LR[0]),
         output_label(pred_DTC[0]) , output_label(predict_RFC[0]), output_label(predict_GBC[0]))
```

```python
#taking the input from user
news_article = str(input())
```

```python
#Predicting the news as genuine or fake by 4 models
manual_testing(news_article)
```
