# Import Dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Data Collection and Preprocessing

```python
movie_dataset = pd.read_csv('/content/movies.csv')
```

```python
movie_dataset.head()
```

```python
# numbers of rows and columns in dataset
movie_dataset.shape
```

```python
#selecting the relevant features for recommendation
selected_features = ["genres"	, "keywords", "tagline","cast", "director"]
print(selected_features)
```

```python
# Replacing the null values with null string for selected features
for feature in selected_features:
  movie_dataset[feature] = movie_dataset[feature].fillna('')
```

```python
#checking for null values
movie_dataset.isnull().sum()
```

```python
# combining all the selected features
combined_features = movie_dataset['genres']+" " +movie_dataset['keywords']+" "+movie_dataset['tagline']+" "+movie_dataset['cast']+" "+movie_dataset['director']
print(combined_features)
```

```python
# convert the textual data to vector feature or simple words in numerical values

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

## Find Similarity Score

```python
#getting bthe similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# Getting the movie name from user
movie_name = input("Enter your Favourite movie name")
```

```python
# Creating the list with all the movie names given in the dataset
list_all_of_the_titles = movie_dataset['title'].tolist()
print(list_all_of_the_titles)
```

```python
# finding the close match for the movie name given  by the user
find_close_match = difflib.get_close_matches(movie_name, list_all_of_the_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
 #find the index of the movie with title

index_of_the_movie = movie_dataset[movie_dataset.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies
similarity_score  = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# Sorting the movies base on their similarity score
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you :  \n ')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_dataset[movie_dataset.index==index]['title'].values[0]
  if (i<31):
    print(i , '.' , title_from_index)
    i+=1
```

# Movie Recommendation System

```python
# Getting the movie name from user
movie_name = input("Enter your Favourite movie name")
# Creating the list with all the movie names given in the dataset
list_all_of_the_titles = movie_dataset['title'].tolist()
# finding the close match for the movie name given  by the user
find_close_match = difflib.get_close_matches(movie_name, list_all_of_the_titles)
close_match = find_close_match[0]
#find the index of the movie with title
index_of_the_movie = movie_dataset[movie_dataset.title == close_match]['index'].values[0]
#getting the list of similar movies
similarity_score  = list(enumerate(similarity[index_of_the_movie]))
# Sorting the movies base on their similarity score
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)
# print the name of similar movies based on the index
print('Movies suggested for you :  \n ')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_dataset[movie_dataset.index==index]['title'].values[0]
  if (i<31):
    print(i , '.' , title_from_index)
    i+=1
```
