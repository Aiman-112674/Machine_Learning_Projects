# Under- Sampling

# Build a sample dataset containing similiar distribution of normal Transacion and Fradulent Tansaction  

# number  of F.T is 492

```python
legit_sample  = legit.sample(n=492 )
```

# Concatenating 2 datasets , Fradulent + legit_ sample

```python
new_dataset = pd.concat([legit_sample , fraud ] , axis = 0)
```

```python
new_dataset.head(5)
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# split the data into Features and Tragets
features -- X  
Targets -- Y


```python
X = new_dataset.drop(columns= 'Class' , axis = 1)
y = new_dataset['Class']
```

```python
print(X)
```

```python
print(y)
```

# Split the data into training and testing data

```python
X_train , X_test , y_train , y_test = train_test_split(X,y, stratify=y ,test_size = 0.2 , random_state = 42)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

# Training the model with training data

```python
model.fit(X_train , y_train)
```

# Evalution of Model based on Accuracy Score

# 1st perform on Training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , y_train)
```

```python
print('Accuracy on Training data : ' , training_data_accuracy)
```

# Accuracy on Test Data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , y_test )
print('Accuracy on Test data :'  , test_data_accuracy)
```
