# Import dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection

```python
cancer_dataset = pd.read_csv("/content/data.csv")
```

# Data Preprocessing

```python
# printing the 1st five rows of dataset
cancer_dataset.head()
```

```python
# printing the last 5 rows of dataset
cancer_dataset.tail(10)
```

```python
# finding the numbers of rows and columns in the dataset
cancer_dataset.shape
```

```python
# finding the information about the dataset
cancer_dataset.info()
```

```python
# finding the missing values
cancer_dataset.isnull().sum()
```

0 -- Benign or not cancerous
1 -- Malignant or cancerous

```python
# Rename 'diagnosis' column to 'target' -- as this is the target column
cancer_dataset.rename(columns={'diagnosis': 'target'}, inplace=True)

# Map M -> 1 (malignant/cancerous), B -> 0 (benign/not cancerous)
cancer_dataset['target'] = cancer_dataset['target'].map({'M': 1, 'B': 0})

# Check the result
cancer_dataset.head()
```

```python
# removing the column no 32 named as Unnamed:32 beacuse it contain Null values NAN
cancer_dataset.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
# Checking the updation
cancer_dataset.head()
```

```python
# finding the Statistical meaures
cancer_dataset.describe()
```

```python
# checking the distribution of the target variable --- Label column
# Finding how many of the data entries are Malignant or Begnin
cancer_dataset['target'].value_counts()
```

```python
# finding the mean of the target column
cancer_dataset.groupby('target').mean()
```

# Splitting the data into Features or Targets

```python
#feature -- Independent Variable
X = cancer_dataset.drop(columns='target' , axis=1)
```

```python
# target  -- Dependent Variable
y = cancer_dataset['target']
```

```python
# Checking the updation
X
```

```python
# Checking the updation
y
```

# Splitting the data into Training data or Testing data

```python
# Train_Test_Split
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size = 0.3 , random_state = 42)
```

```python
# finding the shape of X
X.shape
```

```python
# finding the shape of X_train
X_train.shape
```

```python
# finding the shape of X_test
X_test.shape
```

# Model Creation

```python
# Model Intialization
LR_model = LogisticRegression()
```

# Fitting the data to model

```python
LR_model.fit(X_train , y_train)
```

# Model Evalution

#### Accuracy on Training Data

```python
training_data_prediction = LR_model.predict(X_train)
training_data_accuracy = accuracy_score(y_train , training_data_prediction)
print("Accuracy on Training data : " , training_data_accuracy)
```

#### Accuracy on Testing Data

```python
testing_data_prediction = LR_model.predict(X_test)
testing_data_accuracy = accuracy_score(y_test , testing_data_prediction)
print("Accuracy on Testing data : " , testing_data_accuracy)
```
